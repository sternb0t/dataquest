```python
import pandas as pd
avengers = pd.read_csv("./data/avengers.csv", encoding="iso-8859-1")
```

```python
avengers.head(5)
```

```python
true_avengers = avengers[avengers["Year"] > 1960].copy()
```

```python
true_avengers.shape
```

```python
columns = ['Death1', 'Death2', 'Death3', 'Death4', 'Death5']
```

```python
true_avengers[columns].head(20)
```

```python
def num_deaths(row):
    deaths = 0
    for c in columns:
        if row[c] == "YES":
            deaths +=1
    return deaths
```

```python
deaths = true_avengers.apply(num_deaths, axis=1)
```

```python
true_avengers["Deaths"] = deaths.copy()
```

```python
true_avengers["Deaths"].isnull().value_counts()
```
